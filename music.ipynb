```python
import torch
import torch.nn as nn

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")


X_train = torch.tensor(X_train.values, dtype=torch.float32)
X_test = torch.tensor(X_test.values, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.long)
X_train, X_test = X_train.to(device), X_test.to(device)

y_train = torch.tensor(pd.get_dummies(y_train).values, dtype=torch.float32).to(device)
y_test = torch.tensor(pd.get_dummies(y_test).values, dtype=torch.float32)

y_train_squeeze = np.argmax(y_train.cpu(), axis=1).squeeze()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, roc_auc_score

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dimension is equal to the number of features
D = X_train.shape[1]
C = 10  # Number of classes
H = 100  # Number of hidden units

# Setting hyperparameters
learning_rate = 0.1
lambda_l2 = 1e-3

def build(hidden_layers=1, activation='relu', H=100):

    if activation == 'relu':
        activation = nn.ReLU()
    elif activation == 'sigmoid':
        activation = nn.Sigmoid()
    elif activation == 'tanh':
        activation = nn.Tanh()
    elif activation == 'softplus':
        activation = nn.Softplus()
    else:
        activation = nn.Identity()

    layers = []
    layers.append(nn.Linear(D, H))
    layers.append(nn.BatchNorm1d(H))
    layers.append(activation)
    layers.append(nn.Dropout(p=0.2))

    for _ in range(hidden_layers):
        layers.append(nn.Linear(H, H))
        layers.append(nn.BatchNorm1d(H))
        layers.append(activation)
        layers.append(nn.Dropout(p=0.2))

    layers.append(nn.Linear(H, C))
    
    model = nn.Sequential(*layers)
    model.to(device)
    return model

```

```python
def plot_roc_curves(fpr, tpr, roc_auc, genre):
    lw = 2
    fig, axes = plt.subplots(2, 5, figsize=(16, 8))

    for i, ax in enumerate(axes.flat):
        ax.plot(fpr[i], tpr[i], color='darkorange',
                lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[i])
        ax.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
        ax.set_xlim([-0.1, 1.0])
        ax.set_ylim([-0.1, 1.0])
        ax.set_title(genre[i])

    plt.tight_layout()
    plt.show()


def train(X_train, X_test, y_train, y_train_squeeze, y_test, hidden_layers, activation, H, verbose=False):
    model = build(hidden_layers, activation, H)
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, weight_decay=lambda_l2)
    train_loss = []
    class_labels = np.unique(y_train_squeeze)

    for t in range(1000):
        y_pred = model(X_train)
        #get the predicted class
        score, predicted = torch.max(y_pred, 1)

        predicted = torch.tensor(pd.get_dummies(predicted.cpu()).values, dtype=torch.float32).to(device)
        # Compute and print accuracy
        predicted = np.argmax(predicted.cpu(), axis=1).squeeze()
        correct = (predicted == y_train_squeeze).sum().item()
        accuracy = correct / len(y_train_squeeze)

        loss = criterion(y_pred, y_train)
        optimizer.zero_grad()

        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())

    # Plotting the training loss
    plt.plot(range(1, 1000 + 1), train_loss)
    plt.xlabel("Number of epochs")
    plt.ylabel("Training Loss")
    plt.show()

    with torch.no_grad():
        model.eval()
        y_pred = model(X_test)

        if verbose:
            genre = ['Alternative', 'Anime', 'Blues', 'Classical', 'Country', 'Electronic', 'Hip-Hop', 'Jazz', 'Rap', 'Rock']
            fpr = dict()
            tpr = dict()
            roc_auc = dict()

            for i in range(10):
                fpr[i], tpr[i], _ = roc_curve(y_test.cpu()[:, i], y_pred.cpu()[:, i])
                roc_auc[i] = auc(fpr[i], tpr[i])

            plot_roc_curves(fpr, tpr, roc_auc, genre)

    return roc_auc_score(y_test, y_pred.cpu().detach(), multi_class='ovr', average=None), roc_auc_score(
        y_test, y_pred.cpu().detach(), multi_class='ovr')
```

```python
import random
import torch

seed = 15238783
random.seed(seed)
torch.manual_seed(seed)


rocs, average_roc = train( X_train,X_test, y_train, y_train_squeeze, y_test, 3, 'relu', 100, True)
rocs,average_roc
```
